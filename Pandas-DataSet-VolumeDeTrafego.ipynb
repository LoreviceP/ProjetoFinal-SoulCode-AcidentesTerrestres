```python
#4 - IMPORTANDO O DOCUMENTO E CODIFICANDO DADOS

'''with open("tmp/VolTrafego2019.json") as grades:    
  dados = json.load(grades)'''

df1 = pd.read_json("/tmp/VolTrafego2019.json")
df2 = pd.read_json("/tmp/VolTrafego2020.json")
```

```python
#5 - CONSULTANDO DATAFRAME EM JSON PARA VER SE OS DADOS CARREGARAM DE FORMA CORRETA

df1
```

```python
#6 - NORMALIZANDO DADOS CONTIDOS NOS ARQUIVOS JSON

df1 = pd.json_normalize(df1["empresas_habilitadas_regular"])
df2 = pd.json_normalize(df2["empresas_habilitadas_regular"])
```

```python
#7 - CONSULTANDO TIPO DE DADOS DO DATAFRAME

df1.info()
```

```python
#8 - NORMALIZANDO TIPO DATE NO DATAFRAME1

df1["mes_ano"] = pd.to_datetime(df1["mes_ano"], dayfirst=True)
```

```python
#9 - NORMALIZANDO TIPO DATE NO DATAFRAME2

df2["mes_ano"] = pd.to_datetime(df2["mes_ano"], dayfirst=True)
```

```python
#10 - CONCATENANDO DATAFRAMES

frames = [df1, df2]
df = pd.concat(frames)
```

```python
#11 - VISUALIZANDO DATAFRAME

df
```

```python
#12 - ANALISANDO O TIPO DE DADOS (STRING, INT, FLOAT)

df.dtypes
```

```python
#13 - EDITANDO TIPOS DE DADOS DO DATAFRAME

df.Concessionaria = df.Concessionaria.convert_dtypes(convert_string=True)
df.Sentido = df.Sentido.convert_dtypes(convert_string=True)
df.Praca = df.Praca.convert_dtypes(convert_string=True)
df.Categoria = df.Categoria.convert_dtypes(convert_string=True)
df.Tipo_de_veiculo = df.Tipo_de_veiculo.convert_dtypes(convert_string=True)
df.Volume_total = df.Volume_total.convert_dtypes(convert_integer=True)
```

```python
#14 - ANALISANDO O TIPO DE DADOS (STRING, INT, FLOAT) PARA VER SE AS TROCAS DERAM CERTO

df.info()
```

```python
#16 - CONVERTENDO O VALOR DA COLUNA "Volume Total" DE OBJETO PARA NÚMERICO

df['Volume_total'] = pd.to_numeric(df['Volume_total'],errors = 'coerce')
```

```python
#17 - VERIFICANDO AS TROCAS DOS TIPOS DE DADOS

df.info()
```

```python
#18 - BACKUP DE SEGURANÇA CASO QUALQUER EDIÇÃO SEJA EQUIVOCADA

df = df.copy()
```

```python
#19 - VISUALIZANDO DATAFRAME

df
```

```python
#20 - TROCANDO NOME DA COLUNA "mes_ano" - df.rename(columns={"NomeQueQueroMudar": "NomeNovo"}) + inplace=True

df.rename(columns={"mes_ano": "Data"}, inplace=True)
```

```python
#21 - ANALIZANDO DADOS DAS COLUNAS | COLUNA "Sentido"  - tem de dropar ela, não consta nada de relevante

pd.unique(df["Sentido"])
```

```python
#22 - ANALIZANDO DADOS DAS COLUNAS | COLUNA "Categoria"  - tem de dropar ela, não consta nada de relevante

pd.unique(df["Categoria"])
```

```python
#23 - DROPANDO COLUNAS - APENAS PARA MELHOR VISUALIZAÇÃO DE DADOS MAIS PERTINENTES | UM BACKUP SERA FEITO SOBRE ALGUMAS DESSAS COLUNAS, CASO ELAS NECESSITEM SER CONSULTADAS

df["backup_Categoria"] = df.Categoria

df.drop(["Categoria"], axis=1, inplace=True)
```

```python
#24 - VERIFICANDO DROP / BACKUP

df
```

```python
#25 - DROPANDO COLUNAS SEM RELEVÂNCIA PARA O TRABALHO

df.drop(["backup_Categoria"], axis=1, inplace=True)
```

```python
#26 - VERIFICANDO DROP

df
```

```python
#27 - DIVIDINDO DADOS DA COLUNA DE PRAÇAS PARA VISUALIZAR COMO EXTRAIR OS DADOS

df['Praca'].str.split(" ", expand=True)
```

```python
#28 - DIVIDINDO DADOS DA COLUNA DE PRAÇAS E SALVANDO ELAS EM COLUNAS

df [["1", "2", "BRUF","3","4","5","6","7","8","9","10","11"]] = df["Praca"].str.split(" ", expand=True)
```

```python
#29 - EXCLUINDO COLUNAS RESULTANTES DA DIVISÃO

df.drop(["1"], axis=1, inplace=True)
df.drop(["2"], axis=1, inplace=True)
df.drop(["3"], axis=1, inplace=True)
df.drop(["4"], axis=1, inplace=True)
df.drop(["5"], axis=1, inplace=True)
df.drop(["6"], axis=1, inplace=True)
df.drop(["7"], axis=1, inplace=True)
df.drop(["8"], axis=1, inplace=True)
df.drop(["9"], axis=1, inplace=True)
df.drop(["10"], axis=1, inplace=True)
df.drop(["11"], axis=1, inplace=True)
```

```python
#30 - VISUALIZANDO DATAFRAME

df
```

```python
#31 - DIVIDINDO DADOS DA COLUNA "BRUF" E SALVANDO EM COLUNAS

df [["BR", "UF"]] = df["BRUF"].str.split("/", expand=True)
```

```python
#32 - DROPANDO COLUNAS QUE FICARAM REPETIDAS

df.drop(["BRUF"], axis=1, inplace=True)
df.drop(["Praca"], axis=1, inplace=True)
```

```python
#33 - VISUALIZANDO DATAFRAME

df
```
